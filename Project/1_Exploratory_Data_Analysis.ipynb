```python
#import Libraries
import ast
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset

#Load Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)
```

```python
df_da_us = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_plot=df_da_us['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='darkgrid', palette= 'deep' )
sns.barplot(data=df_plot, x='count', y='job_location', hue='job_location', palette= 'dark:b_r')
plt.title('Top 10 Locations for Data Analyst Jobs in the US')
plt.ylabel('')
plt.xlabel('Number of Job Postings')
plt.show()
```

```python

dict_column = {
    'job_work_from_home': 'work from home offered',
    'job_no_degree_mention': 'degree requirement',
    'job_health_insurance': 'health insurance offered'
    }

fig, ax = plt.subplots(1, 3)   # 1 row, 3 columns
fig.set_size_inches(15, 5)

for i, (x, y) in enumerate(dict_column.items()):
    # get value counts in order [False, True]
    values = df_da_us[x].value_counts().reindex([False, True])
    
    ax[i].pie(
        values,
        labels=['False', 'True'],
        autopct='%1.1f%%',
        startangle=90,
        colors= ['#ff8905', '#87ceeb'],
    )
    ax[i].set_title(y)

plt.show()
```

```python
df_plot=df_da_us['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style='darkgrid', palette= 'deep' )
sns.barplot(data=df_plot, x='count', y='company_name', hue='company_name', palette= 'dark:b_r')
plt.title('Top 10 Companies Hiring Data Analysts in the US')
plt.ylabel('')
plt.xlabel('Number of Job Postings')
plt.show()
```
